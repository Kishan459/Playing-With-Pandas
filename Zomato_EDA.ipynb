```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('/content/zomato.csv',encoding="ISO-8859-1")
ctry_code = pd.read_excel('/content/Country-Code.xlsx')
```

```python
final_df = pd.merge(df,ctry_code,on='Country Code', how = 'left')
```

```python
top_10_restaurants = df['Restaurant Name'].value_counts().reset_index().rename(columns={'index':'restaurant_name','Restaurant Name':'count'}).head(10)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = top_10_restaurants['restaurant_name'],y = top_10_restaurants['count'])
plt.xticks(rotation='vertical')
plt.title('Top 10 most visited Restaurants',fontsize=28)
plt.show();
```

#### top 10 restaurants to frequently visited/placed online order

```python
top_10_cities = final_df['City'].value_counts().reset_index().rename(columns={'index':'city','City':'count'}).head(10)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = top_10_cities['city'],y = top_10_cities['count'],palette='flare')
plt.xticks(rotation='vertical')
plt.title('Top 10 Cities where most transections are happening',fontsize=28)
plt.show();
```

#### Most of the transections are happening in Delhi NCR

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating_count'})
```

```python
ratings['Rating text'].value_counts()
```

```python
ratings_df = ratings.groupby('Rating text')['rating_count'].sum().reset_index().sort_values(by = 'rating_count',ascending=False)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = ratings_df['Rating text'], y = ratings_df['rating_count'])
plt.title('Ratings',fontsize=28)
```


* more then 3500 people are rated avarage.

* and more than 2000 people are not even rated, so we can say that most people prefere not to give ratings

```python
ratings
```

```python
plt.figure(figsize=(15,7))
sns.barplot(ratings['Aggregate rating'],ratings['rating_count'])
plt.show();
```


* from the above visualization we can say that on an avarage most people are rated between 2.9 to 3.9 

* and there are very few people who are rated more than 4

* Not rated count is very high
