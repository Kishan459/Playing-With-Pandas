## find top 5 ted talk which provoked the most oneline discussion

```python
df.sort_values(by='comments').tail()
```

#### there is a problem with this approch because there are some videos which might be there for years and contains more comments then the new one so what we do is we will make a new feature view per comment by dividing views with comments

```python
df['views_per_comments'] = df.views / df.comments 
```

```python
df.sort_values('views_per_comments').head().reset_index()
```

## top 5 ted talk which provoked the most oneline discussion
